```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
GLD2_path = '/media/iap205/Data/Export/cnnimageretrieval-google_landmark_retrieval/trained_network/train_cleaned_test/R101_GLD2_362/google-landmarks-dataset-v2_resnet101_gem_contrastive_m0.85_adam_lr5.0e-07_wd1.0e-04_nnum5_qsize2000_psize22000_bsize5_imsize362/'
GLD2_cln_path = '/media/iap205/Data/Export/cnnimageretrieval-google_landmark_retrieval/trained_network/train_cleaned_test/R101_GLD2_cleaned_362/GLD-v2-cleaned_resnet101_gem_contrastive_m0.85_adam_lr5.0e-07_wd1.0e-04_nnum5_qsize2000_psize22000_bsize5_imsize362/2019-07-01 14:44:17.201905_log.csv'
GLD2_cln_m2_path = '/media/iap205/Data/Export/cnnimageretrieval-google_landmark_retrieval/trained_network/train_cleaned_test/R101_GLD2_cleaned_m2_362/GLD-v2-cleaned-m2_resnet101_gem_contrastive_m0.85_adam_lr5.0e-07_wd1.0e-04_nnum5_qsize2000_psize22000_bsize5_imsize362/2019-07-01 14:44:26.294624_log.csv'
GLD1_resize_path = ''
SfM_120k_path = ''

GLD2 = pd.read_csv(GLD2_path)
GLD2_cln = pd.read_csv(GLD2_cln_path)
GLD2_cln_m2 = pd.read_csv(GLD2_cln_m2_path)
GLD1_resize = pd.read_csv(GLD1_resize_path)
SfM_120k = pd.read_csv(SfM_120k_path)

```

```python
# plot the data
plt.close('all')
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(GLD2['epochs'], GLD2['GLD2_mAP@100'], label='GLD-v2', linewidth=1.0, linestyle='-')
ax.plot(GLD2_cln['epochs'], GLD2_cln['GLD2_mAP@100'], label='GLD-v2-cleaned1', linewidth=1.0, linestyle='-.')
ax.plot(GLD2_cln_m2['epochs'], GLD2_cln_m2['GLD2_mAP@100'], label='GLD-v2-cleaned2', linewidth=1.0, linestyle='--')
ax.plot(GLD1_resize['epochs'], GLD1_resize['GLD2_mAP@100'], label='GLD-v1', linewidth=1.0, linestyle='-')
ax.plot(SfM_120k['epochs'], SfM_120k['GLD2_mAP@100'], label='SfM_120k', linewidth=1.0, linestyle='-')
ax.set(xlabel="epochs", ylabel="mAP@100(%)")
ax.legend(loc='upper left')
ax.grid(axis='y', b=True, which='both', fillstyle='none')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

```
